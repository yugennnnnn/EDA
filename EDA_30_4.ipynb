```python
import pandas as pd

link = 'https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DA0101EN/automobileEDA.csv'

df = pd.read_csv(link)

df.head(10)
```

```python
#correlation

df.corr()
```

```python
df[['price', 'engine-size']].corr()
```

```python
df[['price', 'engine-size', 'horsepower', 'peak-rpm']].corr()
```

```python
#pearson method

from scipy import stats

pearson_coef, p_value = stats.pearsonr(df['price'], df['engine-size'])

print ('pearson_coef is ', pearson_coef, 'p_value is ',  p_value)
```

```python
#group by

df1 = df[['drive-wheels', 'body-style', 'price']]

df1.head(10)
```

```python
df1_groupby = df1.groupby(['drive-wheels', 'body-style'], as_index=False).mean()

df1_groupby
```

```python
#pivot

pivot1 = df1_groupby.pivot(index='drive-wheels' , columns='body-style')
```

```python
pivot1
```

```python
#pivot

pivot2 = df1_groupby.pivot(index='body-style' , columns='drive-wheels')

pivot2
```

```python
#group by

df2 = df[['drive-wheels', 'num-of-doors', 'price']]

df2_groupby = df2.groupby(['drive-wheels', 'num-of-doors'], as_index=False).mean()

df2_groupby
```

```python
#pivot

pivot3 = df2_groupby.pivot(index='num-of-doors' , columns='drive-wheels')

pivot3
```

```python
#heatmap

import numpy as np


fig, ax = plt.subplots()
im = ax.pcolor(pivot3, cmap='RdBu')

#label names
row_labels = pivot3.columns.levels[1]
col_labels = pivot3.index

#move ticks and labels to the center
ax.set_xticks(np.arange(pivot3.shape[1]) + 0.5, minor=False)
ax.set_yticks(np.arange(pivot3.shape[0]) + 0.5, minor=False)

#insert labels
ax.set_xticklabels(row_labels, minor=False)
ax.set_yticklabels(col_labels, minor=False)

#rotate label if too long
plt.xticks(rotation=90)

fig.colorbar(im)
plt.show()
```
